```python
# importing the packages in the environment
import warnings
from transformers import pipeline
from langchain import HuggingFacePipeline
from langchain import LLMChain, PromptTemplate
from transformers import AutoTokenizer, AutoModelForCausalLM

warnings.filterwarnings('ignore')
```

```python
# loading the llama2 model
model = AutoModelForCausalLM.from_pretrained('NousResearch/Llama-2-7b-chat-hf', device_map='auto', load_in_4bit=True)
tokenizer = AutoTokenizer.from_pretrained('NousResearch/Llama-2-7b-chat-hf')
```

```python
# utility method to create the prompt using the system prompt and the instruction
def get_prompt(system_prompt, instruction):
  prompt = "<<SYS>>\n"+system_prompt+" <</SYS>>\n" + '[INST] ' + instruction + '[/INST]'
  return prompt
```

```python
# creating the text-generation pipeline
pipe = pipeline(
    'text-generation',
    model=model,
    tokenizer=tokenizer,
    device_map='auto',
    max_new_tokens=512,
    eos_token_id=tokenizer.eos_token_id
)
```

```python
# creating LLM pipeline from the text-generation pipeline created above
llm = HuggingFacePipeline(pipeline=pipe, model_kwargs={
    'temperature': 0.8
})
```

```python
system_prompt = 'You are an advanced assistent that excels in language translation'
instruction = 'Convert the following text from English to Tamil\n\n## TEXT: {text}\n'
```

```python
template = get_prompt(system_prompt, instruction)
print(template)
```

```python
prompt = PromptTemplate(template=template, input_variables=['text'])
llm_chain = LLMChain(prompt=prompt, llm=llm)
```

```python
text = "Hello, I am Jai!!"
output = llm_chain.run(text).strip()
```
